### Problem 1
Write a code to do an LU factorization of a given matrix with partial pivoting (Algorithm 21.1)

```python
import numpy as np
def LU_factorization(A):
    m = A.shape[0]
    U = A
    P = np.identity(m)
    L = np.identity(m)
    
    for k in range(m-1):
        
        #pivoting
        if np.abs(U[k,k])==0:
            for i in range(k+1,m):
                if np.abs(U[i,k])>np.abs(U[k,k]):
                    # Swaps ith and kth rows to each other
                    U[[k,i]]=U[[i,k]]             
                    L[[k,i]]=L[[i,k]]
                    break

        #LU
        for j in range(k+1,m):
            L[j,k] = (U[j,k] / U[k,k])     
            U[j,k:m] -= np.dot(U[k,k:m],L[j,k])
        
    return P, L, U
```

### Problem 2
Use your code to factor a random 10 Ã— 10 matrix.

```python
A = np.random.random((10,10))
P1, L1, U1 = LU_factorization(A)
#print(P1)
print(L1)
print(U1)
```
